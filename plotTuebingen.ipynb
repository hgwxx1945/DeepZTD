```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
from matplotlib.dates import DateFormatter, DayLocator, HourLocator, drange
fig, axs = plt.subplots(2, 1, figsize=(10, 10))
date1 = datetime.datetime(2022, 6, 25)
date2 = datetime.datetime(2022, 7, 10)
delta = datetime.timedelta(hours=1)
dates = drange(date1, date2, delta)
res=np.load('Tuebingensummer.npy')
v=res[0,:]
e=res[1,:]
gac=res[2,:]
gm=res[3,:]
gnss=res[4,:]
axs[0].plot(dates, v, 'b', label='VMF3')
axs[0].plot(dates, e, 'g', label='ERA5')
axs[0].plot(dates, gnss, 'r', label='GNSS')
axs[0].plot(dates, gac, 'm', label='GACOS')
axs[0].plot(dates, gm, 'k', label='GMLSTM')
axs[0].legend(loc='upper right')
axs[0].set_title('Tuebingen Summer')
axs[0].xaxis.set_major_locator(DayLocator(interval=3))
axs[0].xaxis.set_major_formatter(DateFormatter('%m-%d'))
axs[0].set_xlim(dates[0], dates[-1])
axs[0].yaxis.set_label_text('ZTD (mm)')
#load Tuebingenwinter.npy
res=np.load('Tuebingenwinter.npy')
v=res[0,:]
e=res[1,:]
gac=res[2,:]
gm=res[3,:]
gnss=res[4,:]

#dates
date1 = datetime.datetime(2021, 11, 26)
date2 = datetime.datetime(2021, 12, 11)
delta = datetime.timedelta(hours=1)
dates = drange(date1, date2, delta)

#plot Tuebingenwinter
axs[1].plot(dates, v, 'b', label='VMF3')
axs[1].plot(dates, e, 'g', label='ERA5')
axs[1].plot(dates, gnss, 'r', label='GNSS')
axs[1].plot(dates, gac, 'm', label='GACOS')
axs[1].plot(dates, gm, 'k', label='GMLSTM')
axs[1].legend(loc='upper right')
axs[1].set_title('Tuebingen Winter')
axs[1].xaxis.set_major_locator(DayLocator(interval=3))
axs[1].xaxis.set_major_formatter(DateFormatter('%m-%d'))
axs[1].set_xlim(dates[0], dates[-1])
axs[1].yaxis.set_label_text('ZTD (mm)')
plt.show()
```
